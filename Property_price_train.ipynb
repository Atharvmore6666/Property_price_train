```python
from sklearn.linear_model import LinearRegression
```

```python
LineReg_df= LinearRegression()
```

```python
LineReg_df.fit(df_train_x,df_train_y)
```

```python
pred_train = LineReg_df.predict(df_train_x)
pred_test = LineReg_df.predict(df_test_x)
```

```python
pred_train
```

```python
pred_test
```

```python
# now we gonna apply parameters
```

```python
Rsq= LineReg_df.score(df_train_x, df_train_y)
Rsq
```

```python
# Adj_Rsq  we need to find now
```

```python
df_train_x.shape
N = df_train_x.shape[0]
K = df_train_x.shape[1]
```

```python
Adj_Rsq = 1-(1-Rsq)*(N-1)/(N-K-1)
Adj_Rsq
```
